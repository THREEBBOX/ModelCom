```python
import numpy as np

a = np.loadtxt('Pdata17_1.txt')
print(a.shape)
x = a[:,0::2]
y = a[:,1::2]
x = x.flatten()
y = y.flatten()
print(x.shape," ",y.shape)
```

```python
xy = np.c_[x,y]
d1=[[70,40]]
xy = np.r_[d1,xy,d1]
print(xy.shape)
```

```python
N = xy.shape[0]
t = np.radians(xy)
d = np.array([[6370 * np.arccos(np.cos(t[i, 0] - t[j, 0]) * np.cos(t[i, 1]) * np.cos(t[j, 1]) + np.sin(t[i, 1]) * np.sin(t[j, 1])) for i in
            range(N)] for j in range(N)]).real
```

```python
#d = np.c_[xy,d]
```

```python
print(d.shape)
```

```python
path = np.arange(N)
L =np.inf

```

```python
for i in range(1000):
    path0 = np.arange(1,N-1)#1~100,101和0位置是固定的
    np.random.shuffle(path0)
    path0 =  np.r_[0,path0,N-1]
    L0 = 0
    for i in range(N-1):
        L0 += d[path0[i],path0[i+1]]
    if L0<L:
        path = path0
        L = L0
print(path.shape)
```

```python
import matplotlib.pyplot as plt
xx =[xy[i,0] for i in path]
yy =[xy[i,1] for i in path]
plt.plot(xx,yy,color='r',marker='*')
plt.show()
```
